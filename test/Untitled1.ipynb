```python
import requests
from pprint import pprint
url='https://services.swpc.noaa.gov/json/goes/primary/xrays-3-day.json'
r = requests.get(url)

data=r.json()
time_tags={}
flux={}
for d in data:
    if d['energy'] not in flux:
        flux[d['energy']]=[]
        time_tags[d['energy']]=[]
    flux[d['energy']].append(d['flux'])
    time_tags[d['energy']].append(d['time_tag'])

outputs={'flux':flux, 'time':time_tags}

    

```
